```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
r=pd.read_csv('C:\\Users\\Akshay N R\\Desktop\\Data Science\\projects\\project3\\datasets\\Dataset (1)\\Ratings.csv')
r
```

```python
b=pd.read_csv('C:\\Users\\Akshay N R\\Desktop\\Data Science\\projects\\project3\\datasets\\Dataset (1)\\Books.csv')
b
```

```python
lj = pd.merge(r, b, on='ISBN', how='left')
```

```python
lj
```

```python
lj.isnull().sum()
```

```python
import copy
```

```python
lj1=copy.deepcopy(lj)
```

```python
lj1=lj1.dropna()
```

```python
lj1.isnull().sum()
```

```python
lj1.head()
```

```python
lj1=lj1.rename({'User-ID':'ID','Book-Rating':'BR','Book-Title':'BT','Book-Author':'Author','Year-Of-Publication':'Year'},axis=1)
```

```python
column_order = ['ID', 'ISBN', 'BT','Author','Publisher','Year','BR','Image-URL-S','Image-URL-M','Image-URL-L']
```

```python
lj1 = lj1[column_order]
```

```python
lj1
```

```python
lj1.info()
```

```python
lj1.groupby('BT')['BR'].count().sort_values(ascending=False)[:10]
```

```python
lj1.groupby('Author')['BR'].count().sort_values(ascending=False)[:10]
```

VISUALIZATION

```python
plt.figure(figsize=(6,5))
sns.countplot(x="BR", data=lj1)
plt.title('Rating distribution')
```

```python
plt.figure(figsize=(4,2))
lj1.groupby('BT')['BR'].count().sort_values(ascending=False)[:10].plot.bar()
plt.title('Top Rated Books')
```

```python
plt.figure(figsize=(8,3))
lj1.groupby('Author')['BR'].count().sort_values(ascending=False)[:10].plot.bar()
plt.title('Top Rated Authors')
```

```python
plt.figure(figsize=(10,5))
lj1.groupby('Author')['BR'].count().sort_values(ascending=False)[:10].plot.pie(autopct='%1.1f%%')
plt.title('Top Rated Authors(percent)')
```
